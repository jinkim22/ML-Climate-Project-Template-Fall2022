```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv")
df = df[:-1]
```

```python
co2_df = pd.read_csv("co2_data.csv")
co2_df = co2_df[8035:]
arr = co2_df["CO2"].values
for i in range(len(arr)):
    if arr[i] == 0:
        arr[i] = arr[i-1]
co2_df['CO2'] = arr
```

```python
import matplotlib.pyplot as plt
plt.plot(co2_df["CO2"], df["windspeed"])
plt.ylabel('windspeed')
plt.show()
```
